# Get Company Description

```python
def get_company_description(ticker_symbols):

    stock_info_dict = {}
    for ticker_symbol in ticker_symbols:
      try:
          # URL for Yahoo Finance company profile page
          url = f'https://finance.yahoo.com/quote/{ticker_symbol}/profile?p={ticker_symbol}'

          # Send a GET request to the URL
          response = requests.get(url,headers = {"user-agent":"Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/108.0.0.0 Safari/537.36"})

          # Parse the HTML content of the page
          soup = BeautifulSoup(response.content, 'html.parser')
          # Find the element containing the company description
          #descritpiton_element=soup.find()
          description_element = soup.find('p', {'class': "Mt(15px) Lh(1.6)"})

          # Extract the text of the company description
          stock_info_dict[ticker_symbol] = description_element.text.strip() if description_element else np.nan

      except Exception as e:
          print(f"Error fetching company description for {ticker_symbol}: {str(e)}")
          stock_info_dict[ticker_symbol] = np.nan

    stock_info_dataframe = pd.DataFrame(list(stock_info_dict.items()), columns=['Symbol', 'Long Business Summary'])
    return stock_info_dataframe
```

# Ranking Companies

```python
class Stock:
    def __init__(self, symbol, current_price, earnings_per_share, dividend_yield, debt_to_equity_ratio, growth_rate):
        self.symbol = symbol
        self.current_price = current_price
        self.earnings_per_share = earnings_per_share
        self.dividend_yield = dividend_yield
        self.debt_to_equity_ratio = debt_to_equity_ratio
        self.growth_rate = growth_rate

    def price_to_earnings_ratio(self):
        return self.current_price / self.earnings_per_share

    def evaluate_dividend(self):
        if self.dividend_yield > 0:
            return "This stock pays dividends."
        else:
            return "This stock does not pay dividends."

    def evaluate_debt(self):
        if self.debt_to_equity_ratio < 1:
            return "The company has a low debt-to-equity ratio, which is a positive sign."
        else:
            return "The company has a high debt-to-equity ratio, which may be a concern."

    def estimate_intrinsic_value(self):
        intrinsic_value = self.earnings_per_share * (1 + self.growth_rate) / (self.price_to_earnings_ratio() - self.growth_rate)
        return intrinsic_value

    def investment_potential(self):
        if self.current_price < self.estimate_intrinsic_value():
            return "This stock may be undervalued and has investment potential."
        else:
            return "This stock may be overvalued and may not be a good investment at the current price."


# Example usage:
if __name__ == "__main__":
    stock = Stock("XYZ", 50.0, 4.0, 2.0, 0.5, 0.1)

    print("Stock Symbol:", stock.symbol)
    print("Price-to-Earnings Ratio:", stock.price_to_earnings_ratio())
    print(stock.evaluate_dividend())
    print(stock.evaluate_debt())
    print("Intrinsic Value Estimate:", stock.estimate_intrinsic_value())
    print(stock.investment_potential())
stocks = [
    Stock("XYZ", 50.0, 4.0, 2.0, 0.5, 0.1),
    Stock("ABC", 60.0, 5.0, 1.5, 0.4, 0.12),
    Stock("DEF", 45.0, 3.5, 1.8, 0.6, 0.08),
    # Add more stocks as needed
]

# Create a function to rank stocks by intrinsic value


# Rank the stocks
rank_stocks_by_intrinsic_value(stocks)

# Print the ranked stocks
print("Ranked Stocks by Intrinsic Value:")
for i, stock in enumerate(stocks):
    print(f"{i + 1}. {stock.symbol} - Intrinsic Value: {stock.estimate_intrinsic_value()}")
```
